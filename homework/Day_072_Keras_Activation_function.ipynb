```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# Sigmoid

Sigmoid 數學函數表示方式

sigmoid = lambda x: 1 / (1 + np.exp(-x))


```python
#Sigmoid 數學函數表示方式
#sigmoid = lambda x: 1 / (1 + np.exp(-x))
def sigmoid(x):
    return (1 / (1 + np.exp(-x)))

#Sigmoid 微分
def dsigmoid(x):
    return (sigmoid(x) * (1 - sigmoid(x)))

x = np.linspace(-10,10,1000)

# Draw the grid line in background.
plt.grid()

# 顯現圖示的Title
plt.title('Sigmoid Function')

# 顯現 Sigmoid formula
plt.text(2.5, 0.7, r'$\sigma(x)=\frac{1}{1+e^{-x}}$', fontsize=20)
 
# create the graph
plt.plot(x, sigmoid(x), 'b');
plt.plot(x, dsigmoid(x), 'r');
```

# softmax

```python
#Softmax 數學函數表示方式
def softmax(x):
     return np.exp(x) / float(sum(np.exp(x)))

x = np.linspace(-10,10,1000)

# Draw the grid line in background.
plt.grid()

# 顯現圖示的Title
plt.title('Softmax Function')

# 顯現 Softmax formula
plt.text(0, 0.015, r'$S(y_i)=\frac{e^{y_i}}{\sum_{j}e^{y_{j}}}$', fontsize=20)

# create the graph
plt.plot(x, softmax(x), 'b');
```
