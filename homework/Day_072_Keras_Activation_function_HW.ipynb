# Rectified Linear Unit- Relu 

f(x)=max(0,x)


```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def relu(x):
  return np.array(list(map(lambda x: x if x>0 else 0, x)))

def drelu(x):
  return np.array(list(map(lambda x: 1 if x>0 else 0, x)))

x = np.linspace(-5,5,1000)

# Draw the grid line in background.
plt.grid()

# 顯現圖示的Title
plt.title('Relu Function')
 
# create the graph
plt.plot(x, relu(x), 'b');
plt.plot(x, drelu(x), 'r');
```
